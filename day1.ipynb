```python
f=open('day1_v1_input.txt','r') 
grp_no=1
d={}
grp_sum=0
for i in f:
    if i!='\n':
        grp_sum+=int(i)
    else:
        d[grp_no]=grp_sum
        grp_no+=1
        grp_sum=0

sorted_dict=dict(sorted(d.items(), key=lambda item: item[1],reverse=True))
sorted_keys=list(sorted_dict.keys())
```

```python
highest=sorted_dict[sorted_keys[0]]
print(highest)
```

```python
top3=sorted_keys[:3]
top3_sum=0
for j in top3:
    top3_sum+=sorted_dict[j]

print(top3_sum)
```
